### 1. How to set a custom style on your plot

There are five preseted seaborn themes: 

*   `darkgrid`
*    `whitegrid`
*   `dark`
*  `white`
*  `ticks`

You can change the style of a plot using the **set_style()** method or using the **set()** method with the parameter **style**.

We you use the function **set_style()** it will set the style for the whole sheet and not for just one graph. If you want to use it for only one graph use the with statement before the function.

#### To apply the style to every plot later on

```python
sns.set_style("whitegrid")
plt.figure(figsize=(2,5))
sns.boxplot(y=tips['total_bill'])
```

```python
def customise_boxplot_graph(style, figsize, data):
  sns.set_style(style)
  plt.figure(figsize=figsize)
  sns.boxplot(y=data)
```

```python
customise_boxplot_graph(style='dark', figsize=(2,4), data=tips['total_bill'])
```

### 3. Subplots

Now, what if you want to want to have these figures in a nice overview, next to each other?

You can create a group of subplots.

Say, you want to create a 2 by 2 grid of axes. You can create those 2x2=4 axes, by using the matplotlib `.subplots(r, c, ...)` function, where r is the number of rows, and c the number of columns.

You can also pass arguments that you would pass to `plt.figure()`, such as `figsize`.

```python
# make a 2 by 2 grid of axes with a total figure size of 12 by 12.
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

# Add this figure the the top-left axes (=axes[0][0])
sns.scatterplot(
    x="total_bill", y="tip", data=tips, 
    ax=axes[0][0] # <- Here you specify on which axes you want to show the figure
)

# Add this figure the the top-right axes (=axes[0][1])
sns.barplot(x="day", y="total_bill", data=tips, ax=axes[0][1])

# Add this figure the the bottom-left axes (=axes[1][0])
sns.barplot(x='sex', y='tip', data=tips, palette='muted', ax=axes[1][0])

# Add this figure the the bottom-right axes (=axes[1][1])
sns.barplot(x='sex', y='tip', data=tips, palette=my_own_palette, ax=axes[1][1])

plt.show()
```

If the values on the y-axes are related (for example $), then it's usually a good idea to set their ranges equal to make it easier to compare the two figures.

You can set the range of each y-axis using `ax.set_ylim(ymin, ymax)`, where `ax` are the axes you want to plot in, `ymin` is the lowest value and `ymax` is the highest value on the axis.

To change the range of the x-axis, you guessed it, you would use `ax.set_xlim()`

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.barplot(x="day", y="total_bill", data=tips, ax=axes[0])
axes[1].set_ylim(0, 25)

sns.barplot(x="day", y="tip", data=tips, palette = my_own_palette, ax=axes[1])
axes[0].set_ylim(0, 25)
```

### 4. How to setup titles

In order to set titles and sub-titles you can or use the built-in functions from `seaborn` or use the `matplotlib` packtotal_bill.

Because seaborn is a library built on matplotlib, every manipulation that you can do with matplotlib, you can also apply them to a seaborn graph. 

#### Setup the title

```python
# Set title with matplotlib
sns.barplot(x="day", y="tip", data=tips, palette = my_own_palette)
plt.title('Average tip given on a given day')
plt.show()
```

```python
# Alternative syntax, that can also be used with matplotlib
ax = sns.barplot(x="day", y="tip", data=tips, palette = my_own_palette)
ax.set_title('Average tip given on a given day')
plt.show()
```

Both results look the same. Applying one or the other is a matter of taste more than a matter of functionality.

#### Styling

That's already good but don't you think that the title looks a bit small over there on the top of the graph?

Luckily there is a way to set the title to the position that you want and to the size that you want using the parameters inside the .title() method:
* **loc** : {'center', 'left', 'right'}

* **pad** : The offset of the title from the top of the axes, in points.
* **fontsize**= The size of the font used in the title, 
* **fontweight**="bold" 

```python
# Example if I set the title to the left add a pad between the title and the graph and put a higher font
plt.figure(figsize=(6,6))
sns.barplot(x="day", y="tip", data=tips, palette = my_own_palette)
plt.title('Tip given on a given day', 
          loc = 'left', 
          pad= 20, 
          fontsize=15, 
          fontweight="bold",
         )
plt.show();
```

That's already better. How about a note on where the data came from?

You can use the methods matplotlib's `.text()` to set a figure title and set a text exactly where you want on the graph.

* x :  gives the x location of the start of the text
* y :  gives the y location of the start of the text

```python
# The same figure
plt.figure(figsize=(10, 5))
sns.barplot(x="day", y="tip", data=tips, palette = my_own_palette)
plt.title('Tip given on a given day', 
          loc='left',
          pad=20,  
          fontsize=15, 
          fontweight="bold")

# ...with a little extra
plt.text(s=' Source: Data coming from the Tips dataset in seaborn',
         x = -0.5, 
         y = -0.5, 
         fontsize=9)

plt.show()
```

#### subplots & titles

Let's create a figure with two axes (subplots) and give each a title.

We can give the entire figure a main title, or "super" title using the `.suptitle(text, x=.5 y=1)` function of matplotlib. The position of that title is given by arguments x and y.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.barplot(x="day", y="total_bill", data=tips, ax=axes[0])
axes[0].set_ylim(0, 25)
axes[0].set_title('Total bill on a given day', fontsize=15)

sns.barplot(x="day", y="tip", data=tips, palette=my_own_palette, ax=axes[1])
axes[1].set_ylim(0, 25)
axes[1].set_title('Tip on a given day', fontsize=15)

# Add source description
axes[0].text(s=' Source: Data coming from the Tips dataset in seaborn', x=-.5, y=-5, fontsize=9)

plt.suptitle("Total bill and Tip on a given day", fontsize=20, x=.5, y=1)
plt.show()
```

### 5. Pimp your axis

So this graph looks already nicer than at the beginning but there are still a couple of things that can be tweaked in order to make the best out of it. 

* The legend of the xaxis and the yaxis are too small
* The ticks of the xaxis and yaxis could be more telling

#### Pimp your axis label

Using the .xlabel() and .ylabel() methods you can setup the name of your axis to the value you want.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.barplot(x="day", y="total_bill", data=tips, ax=axes[0])
axes[0].set_ylim(0, 25)
axes[0].set_title('Total bill on a given day', fontsize=15)

sns.barplot(x="day", y="tip", data=tips, palette=my_own_palette, ax=axes[1])
axes[1].set_ylim(0, 25)
axes[1].set_title('Tip on a given day', fontsize=15)

# Add source description
axes[0].text(s=' Source: Data coming from the Tips dataset in seaborn', x=-.5, y=-5, fontsize=9)

plt.suptitle("Total bill and Tip on a given day", fontsize=20)

# Setup axis labels
axes[0].set_xlabel('Day of Week', fontsize = 15)
axes[0].set_ylabel('Total bill (in $)', fontsize = 15)
axes[1].set_xlabel('Day of Week', fontsize = 15)
axes[1].set_ylabel('Tips given (in $)', fontsize = 15)

plt.show()
```

#### Pimp your axis ticks

The ticks on the x-axis are not obvious, and someone not familiar with the database might not know what they represent.

There is a way to change the name associated to the `ticks` of both of your axis using the methods:
* **xticks()**
* **yticks()**

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.barplot(x="day", y="total_bill", data=tips, ax=axes[0])
axes[0].set_ylim(0, 25)
axes[0].set_title('Total bill on a given day', fontsize=15)

sns.barplot(x="day", y="tip", data=tips, palette=my_own_palette, ax=axes[1])
axes[1].set_ylim(0, 25)
axes[1].set_title('Tip on a given day', fontsize=15)

# Add source description
axes[0].text(s=' Source: Data coming from the Tips dataset in seaborn', x=-.5, y=-5, fontsize=9)

plt.suptitle("Total bill and Tip on a given day", fontsize=20)

# Setup axis labels
axes[0].set_xlabel('Day of Week', fontsize = 15)
axes[0].set_ylabel('Total bill', fontsize = 15)
axes[1].set_xlabel('Day of Week', fontsize = 15)
axes[1].set_ylabel('Tips given', fontsize = 15)

for ax in axes:
    # Change the x-ticks
    ax.set_xticks([0,1,2,3])
    ax.set_xticklabels(['Thursday','Friday','Saturday','Sunday'])
    
    # Change the y-ticks
    ax.set_yticks(np.arange(0,25,5))
    ax.set_yticklabels(['${},-'.format(i) for i in np.arange(0,25,5)])

# Bonus: Use subplots_adjust to create more space between the subplots
# plt.subplots_adjust(hspace=.2, wspace=.4, top=.8)
plt.show()
```

### Bonus -  Annotate an insight on the plot

Sometimes adding subtitles and titles is not enough and you would want to be able to point out a specific point on your graph to draw the attention of your audience to the insight that you have found.

For that you can use the .annotate() method also embeded in the matplotlib packtotal_bill
* **xy**: takes the position of the end of the arrow
* **xytext**: takes the position of the beginning of the text
* **arrowprops**: takes parameters for the arrow style

```python
plt.figure(figsize=(8,6))   
sns.barplot(x="day", y="tip", data=tips, palette = my_own_palette)
plt.title('Tip given on a given day', loc = 'left', pad= 30, fontsize=22, fontweight="bold" )
plt.suptitle('Sunday have higher tips', x = 0.745, y = 0.93)
plt.text(s=' Source: Data coming from the Tips dataset in seaborn', x = -0.5, y = -0.5, fontsize = 9)
#annotate a point on your graph
plt.annotate('On Sunday you get higher tips',
             xy=(3,3.2), 
             xytext=(0,3.5),
             arrowprops=dict(arrowstyle="->",color='#233D4D'))
plt.show()
```
